```python
'''matplotlib - subplot'''
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

## creating Subplot using object oriented Programing Method

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
#axes.plot(x,y)
axes
```

```python
# now plot on subplot x,y
fig, axes = plt.subplots(nrows=1,ncols=2)
for i in axes:
    i.plot(x,y)

    
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('1st plot')
axes[1].plot(y,x)
axes[1].set_title('2nd plot')

# if things are overlaping in  or any issue in plot
plt.tight_layout() # or according to my suggestion 
```

## fig size Aspect Ratio and DPI

```python
'''here dpi means dots per inch or pixel per inch'''
fig = plt.figure(figsize=(2,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(5,2))
axes[0].plot(x,y)
axes[0].plot(y,x)
axes[1].plot(y,x)
plt.tight_layout()
```
