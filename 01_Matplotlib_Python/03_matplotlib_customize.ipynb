```python
'''In this Segment we are going to work in Apperence Change colour parameter or chnage 
in apperamce'''
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0,5,11)
y = x **2

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw = 5,alpha = 0.5)  
# either on color we can use rgb # color code 
# linewidth = for change the line width on plot
# alpha = alpha is use for visibility of line 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=2, linestyle='-',marker = 'o', markersize = 5,
            markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'black')  # here we can use  :, steps etc.
```
