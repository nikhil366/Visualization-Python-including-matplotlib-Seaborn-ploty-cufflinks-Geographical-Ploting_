2. matplotlib. pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot    function makes some change to a figure: e.g., creates a figure, creates a plotting area in a          figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# linspace is use to generate a number from one point to another along with there no. of count
x = np.linspace(0,5,11)
y = x ** 2
print(x,y)
```

```python
'''normal plot  via matplotlib'''
plt.plot(x,y,'r')
#r for color
# if you are not using jupyter then use plt.show()
plt.xlabel('x_lable')
plt.ylabel('y_lable')
plt.title('Normal_plot')
```

## Subplot
Subplot is use for plot the multiple graph at one place and we can do it via the subplot.

```python
'''ploting subplot
here im subplot 1st argument for row and 2nd for column and third for display numbera '''

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
'''ploting 3 subplot in 1 row'''
plt.subplot(1,3,1)
plt.plot(x,y,'r')

plt.subplot(1,3,2)
plt.plot(x,y,'b')

plt.subplot(1,3,3)
plt.plot(x,y,'g')
```

## Matplotlib Object oriented API Method

```python
'''here we use object oriented and create a object of plt'''
fig = plt.figure() # figure is a function of pyplot which is use to plot a blank graph
print(fig)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X__bar')
axes.set_ylabel('Y__bar')
axes.set_title('bar')
```

```python
'''
here in this axes1 [0.1,0.1,0.8,0.8] 
(1) - 0.1 is 10 percent start from left
(2) - 0.1 is 10 percent from bottom
(3) - 0.8 is for canvas size and width
(4) - 0.8 is for canvas size and height

here in this axes2 [0.2,0.5,0.4,0.3]
(1) - 0.2 is 10 percent start from left
(2) - 0.5 is 10 percent from bottom
(3) - 0.4 is for canvas size and width
(4) - 0.3 is for canvas size and height

'''
```

```python

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

```
