```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Graph')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y)
ax.plot(x,y)
ax.set_title('Graph')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.tight_layout()
```
