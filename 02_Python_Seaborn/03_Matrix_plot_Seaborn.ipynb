# Pivot
### A pivot table is a table of statistics that summarizes the data of a more   extensive table. This summary might include sums, averages, or other        statistics, which the pivot table groups together in a meaningful way. Pivot tables are a technique in data processing.
### here we have flight data but the data are not in the form of matrix that we want so for matrix we have to convert data into pivot for matrix data.


```python
flights.head()
```

```python
# converting into pivot
flight_pivot = flights.pivot_table(values='passengers',index='month',columns='year')
flight_pivot
```

```python
sns.heatmap(flight_pivot)
```

```python
'''So for viusalize data more clearly we can cmap(colormap) and distungish then with linecolor and decide width according to requirements'''
sns.heatmap(flight_pivot,cmap='magma',linecolor='black',linewidths=2)
```

# Clustermap
## The clustermap uses hierarchal clustering to produce a clustered version of the heatmap
## basically cluster map is using for cluster same group into same scenario for better understand see downside example.

```python
sns.clustermap(flight_pivot)
```

```python
# More options to get the information a little clearer like normalization
sns.clustermap(flight_pivot,cmap='coolwarm',standard_scale=1)
```
