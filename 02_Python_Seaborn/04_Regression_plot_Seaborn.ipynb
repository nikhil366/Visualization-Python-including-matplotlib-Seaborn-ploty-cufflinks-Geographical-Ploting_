## Linear model with Seaborn

```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
# linear plot
sns.lmplot(x = 'total_bill', y = 'tip', data= tips)
```

```python
# linear plot with hue (hue is used for sepration on the basis of categorical)
sns.lmplot(x='total_bill', y = 'tip', data= tips, hue = 'sex', markers = ['*','+'])
```

```python
# now suppose if we want to plot a linear plot on the basis of sex then we can do it via this method
sns.lmplot(x = 'total_bill', y = 'tip' , data = tips, col= 'smoker')
```

```python
# now suppose if we want to plot a linear plot on the basis of sex then we can do it via this method
sns.lmplot(x = 'total_bill', y = 'tip' , data = tips, col= 'sex', row= 'time')
```
