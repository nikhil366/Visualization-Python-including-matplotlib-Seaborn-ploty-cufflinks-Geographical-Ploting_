## Categorical Plot
Here we will discuss about Categorical plot. Categorical plot or Categorical data are Categorical variables represent types of data which may be divided into groups. Examples of categorical variables are race, sex(Male / Female), age group, and educational level(intermediate / Matric / Graduate).  

```python
import seaborn as sns 
import numpy as np
%matplotlib inline
```

```python
'''In this data set categorical data is sex, smoker, day, time these all are categorical data type
'''
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
# by default it will show you mean of total_ bill but we can use estimator function to change another statistical value or aggregate function  for example std.
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
'''An aggregate function performs a calculation on a set of values, and returns a single value. Except for COUNT(*) , aggregate functions ignore null values. Aggregate functions are often used with the GROUP BY clause of the SELECT statement. All aggregate functions are deterministic'''
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

## Countplot
Basically count plot is use to find out count of variable for example we want count of male and female then we can find out via countplot.

```python
# here we can say that males are more then female in data set
sns.countplot(x='sex',data=tips)
```

## Boxplot 
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are.

To completly understand how an boxplot work and where is the min, max, quartile are place we can see in next image that uploded

```python
sns.boxplot(x = 'day' ,y= 'total_bill' ,data =tips)
```

```python
# Hue with boxplot here with the help of hue we can understand more deeply boxplot on the basis of categorical data.
sns.boxplot(x = 'day' ,y= 'total_bill' ,data =tips, hue = 'sex' )
```

```python
sns.boxplot(x = 'day' ,y= 'total_bill' ,data =tips, hue = 'smoker' )
```

```python
# violinplot    
sns.violinplot(x = 'day' ,y= 'total_bill' ,data =tips, hue = 'smoker' )
```

## Stripplot
stripplot is like as scatter plot 

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# here in stripplot data points are merging to each other so for remove this noise we can use jitter with the help of jitter you can see the points more clearly
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter= True)
```

```python
# here in strip plot we can use function hue (hue use to distungish data point according to data)  and we can use split function to split the data. 
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter= True, hue='sex', split = True)
```
