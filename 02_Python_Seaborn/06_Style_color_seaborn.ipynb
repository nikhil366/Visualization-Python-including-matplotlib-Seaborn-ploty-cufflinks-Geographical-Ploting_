```python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
sns.set_style('white')
sns.countplot(x = 'sex',data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x = 'sex',data = tips)
```

```python
# for dark background
sns.set_style('whitegrid')
sns.countplot(x = 'sex',data = tips)
```
