## Dist plot 
Dist plot is Distribution of univariate set of observation. 
univariate means - one Variable

```python
# dist plot
'''here distplot apply on one feature and we apply dist plot on total_bill
and now we get how much our price is vary'''
# here kde is Kernel density estimation
'''A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions'''
sns.displot(tips['total_bill'], kde=True, bins = 10)
```

## Jointplot
In jointplot we can combine 2 univariate data into a bivariate data. Basically its mean we can plot a distribution of more then two features.
Bivariate = two Variables

```python
'''here we can compare two things two each other with scatterplot'''
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
'''here we can compare two things two each other with hexplot'''
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
'''here we can compare two things two each other with regression with scatter plot '''
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
'''here we can compare two things two each other with kde basically kde is show 2 dimension that where density of points are matches the most'''
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

## Pairplot
pairplot can work for both catagorical and for numerical column.
pairplot show relationship accross an entire data frame.

```python
sns.pairplot(tips)
```

```python
'''here we use categorical value for our numeric data'''
sns.pairplot(tips, hue='sex')
```

```python
'''here palette is use for change in color for plots'''
sns.pairplot(tips, hue='sex',palette='coolwarm')
```
